```python
# Global variables to store books and users
books = {}  # Dictionary to store books and their availability status
users = []  # List to store registered users

# Function to add a book
def add_book(book_title):
    if book_title in books:
        print("Book already exists.")
    else:
        books[book_title] = True  # True means the book is available
        print(f"Book '{book_title}' added successfully.")

# Function to remove a book
def remove_book(book_title):
    if book_title in books:
        del books[book_title]
        print(f"Book '{book_title}' removed successfully.")
    else:
        print("Book not found.")

# Function to search for a book
def search_book(book_title):
    if book_title in books:
        status = "available" if books[book_title] else "checked out"
        print(f"Book '{book_title}' is {status}.")
    else:
        print("Book not found.")

# Function to check out a book
def check_out_book(book_title, user):
    if book_title in books:
        if books[book_title]:
            books[book_title] = False
            print(f"Book '{book_title}' checked out by {user}.")
        else:
            print("Book is already checked out.")
    else:
        print("Book not found.")

# Function to check in a book
def check_in_book(book_title):
    if book_title in books:
        if not books[book_title]:
            books[book_title] = True
            print(f"Book '{book_title}' checked in successfully.")
        else:
            print("Book was not checked out.")
    else:
        print("Book not found.")

# Function to register a user
def register_user(user):
    if user in users:
        print("User already registered.")
    else:
        users.append(user)
        print(f"User '{user}' registered successfully.")

# Function to remove a user
def remove_user(user):
    if user in users:
        users.remove(user)
        print(f"User '{user}' removed successfully.")
    else:
        print("User not found.")

# Function to display all books
def display_books():
    if books:
        print("Books in the library:")
        for book, status in books.items():
            status_str = "available" if status else "checked out"
            print(f"- {book} ({status_str})")
    else:
        print("No books in the library.")

# Function to display all users
def display_users():
    if users:
        print("Registered users:")
        for user in users:
            print(f"- {user}")
    else:
        print("No registered users.")

# Main program
def main():
    while True:
        print("\nLibrary Management System")
        print("1. Add Book")
        print("2. Remove Book")
        print("3. Search Book")
        print("4. Check Out Book")
        print("5. Check In Book")
        print("6. Register User")
        print("7. Remove User")
        print("8. Display Books")
        print("9. Display Users")
        print("10. Exit")

        try:
            choice = int(input("Enter your choice: "))
        except ValueError:
            print("Invalid input. Please enter a number between 1 and 10.")
            continue

        if choice == 1:
            book_title = input("Enter book title: ")
            add_book(book_title)
        elif choice == 2:
            book_title = input("Enter book title: ")
            remove_book(book_title)
        elif choice == 3:
            book_title = input("Enter book title: ")
            search_book(book_title)
        elif choice == 4:
            book_title = input("Enter book title: ")
            user = input("Enter your name: ")
            check_out_book(book_title, user)
        elif choice == 5:
            book_title = input("Enter book title: ")
            check_in_book(book_title)
        elif choice == 6:
            user = input("Enter user name: ")
            register_user(user)
        elif choice == 7:
            user = input("Enter user name: ")
            remove_user(user)
        elif choice == 8:
            display_books()
        elif choice == 9:
            display_users()
        elif choice == 10:
            print("Exiting the system.")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 10.")

if __name__ == "__main__":
    main()
```
