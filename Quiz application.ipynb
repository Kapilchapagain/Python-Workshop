```python
import time

# List of questions, choices, and answers (each question is a tuple)
questions = [
    ("What is the capital city of Spain?", ["a) Madrid", "b) London", "c) Lisbon", "d) Berlin"], "a"),
    ("Which country won the Fifa World Cup 2022?", ["a) Portugal", "b) France", "c) Germany", "d) Argentina"], "d"),
    ("Which is the only anime to win the Oscar Award?", ["a) Naruto", "b) Spirited Away", "c) One Piece", "d) Bleach"], "b"),
    ("Who is the author of the 'Harry Potter' series?", ["a) J.K. Rowling", "b) Philip Pullman", "c) C.S. Lewis", "d) Rudyard Kipling"], "a"),
    ("Which river is the longest in the world?", ["a) Amazon River", "b) Nile River", "c) Yangtze River", "d) Mississippi River"], "b"),
    ("Who directed the movie 'Inception'?", ["a) Steven Spielberg", "b) Martin Scorsese", "c) Christopher Nolan", "d) Quentin Tarantino"], "c"),
    ("Which artist is known for the song 'Shape of You'?", ["a) Taylor Swift", "b) Ed Sheeran", "c) Bruno Mars", "d) Justin Bieber"], "b")
    ("Who was the first President of the Nepal?", ["a) George Washington", "b) Ram Baran Yadav", "c) Abraham Lincoln", "d) Kp Oli"], "a"),
    ("What is the value of Pi (π) to two decimal places?", ["a) 3.12", "b) 3.14", "c) 3.16", "d) 3.18"], "b"),
    ("What is the square root of 64?", ["a) 6", "b) 7", "c) 8", "d) 9"], "c"),
    ("In Greek mythology, who is the god of the sea?", ["a) Zeus", "b) Hades", "c) Poseidon", "d) Apollo"], "c"),
    ("Which Norse god wields the hammer Mjölnir?", ["a) Odin", "b) Thor", "c) Loki", "d) Frey"], "b"),
    ("In which sport would you perform a slam dunk?", ["a) Baseball", "b) Basketball", "c) Tennis", "d) Football"], "b"),
    ("Who painted the Mona Lisa?", ["a) Vincent van Gogh", "b) Pablo Picasso", "c) Leonardo da Vinci", "d) Michelangelo"], "c"),
    ("The Starry Night is a famous painting by which artist?", ["a) Claude Monet", "b) Vincent van Gogh", "c) Pablo Picasso", "d) Salvador Dalí"], "b"),
    ("Which band released the album 'Abbey Road'?", ["a) The Rolling Stones", "b) The Beatles", "c) Led Zeppelin", "d) Pink Floyd"], "b"),
    ("Who is known as the 'King of Pop'?", ["a) Elvis Presley", "b) Michael Jackson", "c) Freddie Mercury", "d) Prince"], "b"),
    ("Which language has the most native speakers worldwide?", ["a) English", "b) Spanish", "c) Mandarin Chinese", "d) Hindi"], "c"),
    ("What is the official language of Brazil?", ["a) Spanish", "b) French", "c) Portuguese", "d) English"], "c"),
    ("Which is the smallest planet in our solar system?", ["a) Venus", "b) Mars", "c) Mercury", "d) Neptune"], "c"),
]

# Function to save progress to a file
def save_progress(name, score):
    with open("quiz_progress.txt", "a") as file:
        file.write(f"{name}: {score} points\n")

# Function for handling quiz
def quiz():
    score = 0
    total_questions = len(questions)
    
    # Input the user's name
    name = input("Enter your name: ")
    
    print("\nWelcome to the quiz, {}! Answer the following questions:\n".format(name))
    
    # Tracking time
    track_time = time.time()
    
    # Loop through the questions
    for i, (question, options, correct_answer) in enumerate(questions):
        print(f"Q{i + 1}: {question}")
        for option in options:
            print(option)
        
        # Get user's answer
        answer = input("Your answer (a/b/c/d): ").lower().strip()
        
        # Exception handling for invalid input
        if answer not in ['a', 'b', 'c', 'd']:
            print("Invalid choice! Please enter a, b, c, or d.\n")
            continue
        
        # Check if the answer is correct or not
        if answer == correct_answer:
            print("Correct!\n")
            score =score + 1
        else:
            print(f"Incorrect! The correct answer is {correct_answer}.\n")
    
    # Time taken to complete the quiz
    end_time = time.time()
    time_taken = end_time - track_time
    
    # Typecasting and displaying results
    print(f"\n{name}, you've completed the quiz!")
    print(f"Your score: {score}/{total_questions}")
    print(f"Time taken: {int(time_taken)} seconds")
    
    # Save progress
    save_progress(name, score)

# Run the quiz
try:
    quiz()
except Exception as e:
    print(f"An error occurred: {e}")
```
